# 서울특별시_대여소별 공공자전거 대여가능 수량(1시간 단위) 번역 코드

```python
import pandas as pd

first_df = pd.read_csv('./data/서울특별시_대여소별 공공자전거 대여가능 수량(1시간 단위)_2023년도_1분기/data_230201~230215.csv', encoding='cp949')
```

```python
print(first_df.columns)
```

```python
first_df.columns = ['Date', 
              'Rental_location_number', 
              'Name_of_the_rental_location', 
              'Time',
              'Num_of_available bikes']
```

```python
first_df
```

```python
second_df = pd.read_csv('./data/서울특별시_대여소별 공공자전거 대여가능 수량(1시간 단위)_2023년도_1분기/data_230216~230228.csv', encoding='cp949')
```

```python
second_df.columns = ['Date', 
              'Rental_location_number', 
              'Name_of_the_rental_location', 
              'Time',
              'Num_of_available bikes']
```

```python
second_df
```

```python
result = pd.concat([first_df, second_df], axis=0)
```

```python
result
```

```python
result.to_csv('./data/available_bikes/Num_of_available_bikes_02.csv', encoding='utf-8', index=False)
```
